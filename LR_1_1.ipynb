```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv("/content/drive/MyDrive/Folder_dataset/games.csv") #читаем файл, убираем столбец индексов
dataset.head()
```

```python
df = dataset.head(1000) #оставил 1000 строк
df
```

```python
df.isnull().sum() #количество пропущенных значений в каждом столбце
```

```python
df["total_points"] = df["PTS_home"] + df["PTS_away"]
df.head()
```

```python
sns.set_style("darkgrid")
ax = sns.boxplot(x = "SEASON", y = "total_points", hue="HOME_TEAM_WINS", data = df) # ящик с усами (распределение и выбросы)
```

```python
hist = df.plot.hist(y = "total_points", bins = 20, grid=False, color='#86bf91') # гистограмма
hist.set_title("Частота набора определённого количества очков") # название гистограммы
hist.set_xlabel("Количество очков", labelpad=20, weight='bold', size=12) # Название оси Х
hist.set_ylabel("Количество матчей", labelpad=20, weight='bold', size=12) # Название оси Y
```

```python
df["GAME_STATUS_TEXT"] = df["GAME_STATUS_TEXT"].fillna("Final") # Заполнить значение в таблице конкретным значением
df["HOME_TEAM_ID"] = df["HOME_TEAM_ID"].fillna("Missing value") # Заполнить значение в таблице конкретным значением
df["TEAM_ID_away"] = df["TEAM_ID_away"].fillna("TEAM_ID_away") # Заполнить значение в таблице конкретным значением
df['PTS_home'] = df['PTS_home'].fillna(df['PTS_home'].mean()) # Заполнить значение в таблице средним
df['AST_home'] = df['AST_home'].fillna(df['AST_home'].mean()) # Заполнить значение в таблице средним
df['total_points'] = df['total_points'].fillna(df['total_points'].mean()) # Заполнить значение в таблице средним
df["GAME_DATE_EST"] = df["GAME_DATE_EST"].bfill() # Заполнить значение в таблице предыдущим значением
df["SEASON"] = df["SEASON"].bfill() # Заполнить значение в таблице предыдущим значением
df.head(5)
df.isnull().sum() #количество пропущенных значений в каждом столбце
```

```python
Q1 = df['total_points'].quantile(0.25)
Q3 = df['total_points'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_df = df[(df['total_points'] >= lower_bound) & (df['total_points'] <= upper_bound)]
filtered_df
```

```python
opp_filtered_df = df[(df['total_points'] <= lower_bound) | (df['total_points'] >= upper_bound)]
opp_filtered_df
```

```python
import numpy as np
games_info = pd.pivot_table(df, index = "SEASON", values = ["total_points", "FG_PCT_home", "FG3_PCT_home", "AST_home", "REB_home"], columns = "HOME_TEAM_WINS", aggfunc=np.mean)
games_info # сводная таблица
```
